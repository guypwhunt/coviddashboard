```python
# Import libraries
from uk_covid19 import Cov19API
import json

# Define how you want the API to filter the COVID data
filters = [
    'areaType=overview' # note each metric-value pair is inside one string
]

# Define the structure of the data you want retruned from the API
structure = {
    "date": "date",
    "cases": "newCasesByPublishDate",
    "hospital": "newAdmissions",
    "deaths": "cumDeaths28DaysByDeathDateRate"    
}

# Create an API object
api = Cov19API(filters=filters, structure=structure)

# Call the API -- this needs exception handling
timeseries=api.get_json()

# Open timeseries file with write priveledges and write api output to
with open("timeseries.json", "wt") as OUTF:
    json.dump(timeseries, OUTF)

# Import the data manipulation and data visulisation libraries
import pandas as pd
import matplotlib.pyplot as plt

# Embed the output of matplotlib output in dashboard
%matplotlib inline

# Make the figures larger
plt.rcParams['figure.dpi'] = 100

# Open timeseries JSON file with read priveledges
with open("timeseries.json", "rt") as INFILE:
    data=json.load(INFILE)

# Extract the data out of the JSON file
datalist=data['data']

# Extract all dates from the datalist and sort them in order
dates=[dictionary['date'] for dictionary in datalist ]
dates.sort()

# Function to create a panda date from the date string
def parse_date(datestring):
    """ Convert a date string into a pandas datetime object """
    return pd.to_datetime(datestring, format="%Y-%m-%d")

# Create a panda date for the first date in the series
startdate=parse_date(dates[0])

# Create a panda date for the last date in the series
enddate=parse_date(dates[-1])

# Create a list of dates from the start date to the end date
index=pd.date_range(startdate, enddate, freq='D')

# Create a timeseries of dates with the columns
timeseriesdf=pd.DataFrame(index=index, columns=['cases', 'hospital', 'deaths'])

# Update the timeseries with data for the columns ['cases', 'hospital', 'deaths']
for entry in datalist: # each entry is a dictionary with date, cases, hospital and deaths
    date=parse_date(entry['date'])
    for column in ['cases', 'hospital', 'deaths']:
        # check that nothing is there yet - just in case some dates are duplicated,
        # maybe with data for different columns in each entry
        if pd.isna(timeseriesdf.loc[date, column]): 
            # replace None with 0 in our data 
            value= float(entry[column]) if entry[column]!=None else 0.0
            # this is the way you access a specific location in the dataframe - use .loc
            # and put index,column in a single set of [ ]
            timeseriesdf.loc[date, column]=value
            
# Fill in any empty values with 0s
timeseriesdf.fillna(0.0, inplace=True)

# Plot the timeseries data
timeseriesdf.plot()

# Plot the timeseries data in log scale
timeseriesdf.plot(logy=True)

# Save the timeseriesdf to a pickle file -- this could be used as the original data source
timeseriesdf.to_pickle("timeseriesdf.pkl")
```

```python

# This is the same as above but for a different dilter and structure
from uk_covid19 import Cov19API
import json

# Define how you want the API to filter the COVID data
filters = [
    'areaType=nation',
    'areaName=England'
]


# Define the structure of the data you want retruned from the API
structure = {
    "males": "maleCases",
    "females": "femaleCases"
}

# Create an API object
api = Cov19API(filters=filters, structure=structure)

# Call the API -- this needs exception handling
agedistribution=api.get_json()

# Open agedistribution file with write priveledges and write api output
with open("agedistribution.json", "wt") as OUTF:
    json.dump(agedistribution, OUTF)

# Open agedistribution file with read priveledges and extract data
with open("agedistribution.json", "rt") as INFILE:
    data=json.load(INFILE)

# Extract data from JSON
datadic=data['data'][0] # data['data'] is a list

# Create a list of male's data
males=datadic['males']

# Create a list of female's data
females=datadic['females']

# Create a list of age ranges
ageranges=[x['age'] for x in males] # each entry of males is a dictionary

# Create function to return the lowest value in the age range
def min_age(agerange):
    agerange=agerange.replace('+','') # remove the + from 90+
    start=agerange.split('_')[0]
    return int(start)

# Sort the age range using the lowest value in the age range
ageranges.sort(key=min_age)

# Create an age data frame with the age range and column headers
age_df=pd.DataFrame(index=ageranges, columns=['males','females', 'total'])

# Fill in the dataframe values for males
for entry in males: # each entry is a dictionary
    ageband=entry['age'] # our index position
    age_df.loc[ageband, 'males']=entry['value']
    
# Fill in the dataframe values for females
for entry in females:
    ageband=entry['age']
    age_df.loc[ageband, 'females']=entry['value']
    
# # Fill in the dataframe values for totals
age_df['total']=age_df['males']+age_df['females']

# Plot the dataframe for males and females
age_df.plot(kind='bar', y=['males','females'])

# Plot the dataframe for total
age_df.plot(kind='bar', y='total')

# Save the age_df to a pickle file -- this could be used as the original data source
age_df.to_pickle("agedf.pkl")

```

```python
# This involves making the graphs interactive

# Import the neccessary libraries
import ipywidgets as wdg
import pandas as pd
import matplotlib.pyplot as plt

# Embed matplot output in dashboard
%matplotlib inline

# Make figures larger
plt.rcParams['figure.dpi'] = 100

# Define a function to call the Covid API, this function will be called when someone presses the button
def access_api(button):
    # Define how you want the API to filter the COVID data
    filters = ['areaType=overview']

    # Define the structure of the data you want retruned from the API
    structure = {"date": "date", "cases": "newCasesByPublishDate", "hospital": "newAdmissions","deaths": "cumDeaths28DaysByDeathDateRate"}

    # Create an API object
    api = Cov19API(filters=filters, structure=structure)

    # Call the API -- this needs exception handling
    timeseries=api.get_json()

    # Open timeseries file with write priveledges and write api output to
    with open("timeseries.json", "wt") as OUTF:
        json.dump(timeseries, OUTF)

    print("I'm downloading data from the API...")
    print("...all done.")
    
# see the doc for the parameters    
apibutton=wdg.Button(
    description='Refresh data',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click to download current Public Health England data',
    icon='download' # (FontAwesome names without the `fa-` prefix)
)

# register the callback function with the button
apibutton.on_click(access_api)

# this is an iPython function that generalises print for Jupyter Notebooks; we use it to 
# display the widgets
display(apibutton)
```
